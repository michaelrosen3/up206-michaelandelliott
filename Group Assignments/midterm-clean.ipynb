We are looking specifically at potted trees, which are most common in places where the topography (flat, gridded) supports them. And slicing it along these dimensions, it is clear North Oakland is the champion of street trees.

But wait! We don't have to simply rely on our eyes — we can use statistics to prove that high numbers of street trees and low numbers of street trees cluster near one another. We'll start by importing the necessary libraries and dropping all rows with null values in order for the analysis to work correctly.

```python
import esda
from esda.moran import Moran, Moran_Local

import splot
from splot.esda import moran_scatterplot, plot_moran, lisa_cluster,plot_moran_simulation

import libpysal as lps

oak_trees_special=oak_trees_income.dropna()
```

Next, we will assign spatial weights to our data. Because there are relatively few census blocks, we assigned the five nearest blocks as the spatial neighbors for our analysis. After this, we'll create a new column in our dataframe to hold the spatial weight values.

```python
wq = lps.weights.KNN.from_dataframe(oak_trees_special,k=5)
wq.transform = 'r'
oak_trees_special['trees in census block lag'] = lps.weights.lag_spatial(wq, oak_trees_special['trees in census block'])
```

We will now calculate our global Moran I statistic for trees in census blocks and then create a scatterplot to be able to visually process the results.

```python
y = oak_trees_special['trees in census block']
moran = Moran(y, wq)
moran.I
```

```python
fig, ax = moran_scatterplot(moran, aspect_equal=True)
plt.show()
```

Sweet! As predicted, there is actually a pretty strong association between the quantity of trees in a census block and the number of trees in the surrounding neighborhoods, or census blocks. This lends credence to our thesis that street trees cluster in predictable patterns. And by running the p.sim command, we can see that is a statistically significant relationship.

```python
moran.p_sim
```

But we're not done! Now we're going to run a local spatial autocorrelation to determine where we see high levels of both street trees and spatial lag, as well as low levels of street trees and spatial lag.

```python
lisa = esda.moran.Moran_Local(y, wq)
fig,ax = plt.subplots(figsize=(20,15))
moran_scatterplot(lisa, ax=ax, p=0.05)
ax.set_xlabel("street trees")
ax.set_ylabel('spatial lag street trees')

# add some labels
plt.text(1.95, 0.5, "HH", fontsize=25)
plt.text(1.95, -1, "HL", fontsize=25)
plt.text(-2, 1, "LH", fontsize=25)
plt.text(-1, -1, "LL", fontsize=25)
plt.show()
```

Let's see what that looks like on a map.

```python
fig, ax = plt.subplots(figsize=(14,12))
lisa_cluster(lisa, oak_trees_special, p=0.05, ax=ax)
plt.show()
```

Well, well, well. Our local spatial autocorrelation analysis shows high/high clusters at the northernmost boundary of Oakland, as well as directly below the exclusionary suburb of Piedmont. On the flip side, we see low-low clusters across the flatlands of East Oakland, as well as a substantial low-low cluster (curiously) in the far east hills.

These results buttress what we already saw on our more simplified map of street trees by census blocks. The darkest parts of the map — in other words, in the census blocks where the most street trees are present — are clustered in North Oakland. Here's a zoomed-in map showing the top five census blocks by street trees, mapped onto a base map.

```python
oak_sorted=oak_trees_income.sort_values(by='trees in census block',ascending=False)
oak_sorted=oak_sorted.head(5)
oak_sorted_webmercator=oak_sorted.to_crs(epsg=3857)

ax = oak_sorted_webmercator.plot(color="green", figsize=(12, 12))
plt.title(label="Top 5 Oakland census blocks by street trees",
          fontsize=25,
          color="black")
ctx.add_basemap(ax, crs=oak_sorted_webmercator.crs.to_string())
```

As you can see, all five census blocks are clustered together in one specific part of the city.
